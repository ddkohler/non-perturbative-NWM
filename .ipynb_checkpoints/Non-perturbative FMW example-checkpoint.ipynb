This notebook will try to apply an electric field to a full system hamiltonian and try to extract a subset of the total polarization that eminates in the FWM direction of interest.  There will be no assumptions about the k-vector additions in which the electric field interacts--all interactions will be considered.  This method was developed by the Domcke group for 2 beam experiments ("Nonperturbative approach to femtosecond spectroscopy:  General theory and application to multidimensional nonadiabatic photoisomerization processes"), and by the Fleming group for three beam experiments and higher (Two-dimensional optical three-pulse photon echo spectroscopy. I. Nonperturbative approach to the calculation of spectra).  

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 1.5 * matplotlib.rcParams['savefig.dpi']
```

## Part 1:  Define the hamiltonian

We will first assume a 3-level system, with equal energy spacings between levels ($\omega_{10} = \omega_{21}$).  Three states means $3^2$ density matrix elements will be needed to define the entire hamiltonian.  We will define this system by the vector:
\begin{equation}
\rho = \begin{bmatrix} 00 \\ 10 \\ 20 \\ 01 \\ 02 \\ 11 \\ 21 \\ 12 \\ 22 \end{bmatrix}
\end{equation}
The vector will be propagated according to the liouville equation in the slowly-varying envelope representation ($\tilde{\rho}_{ij} = \rho_{ij}e^{-i\omega_{ij}t}$):
\begin{equation}
\begin{aligned}
    \frac{d}{dt}\tilde{\rho}_{ij} =& -\Gamma_{ij} \tilde{\rho}_{ij} \\
      & + \overbrace{\frac{i}{2\hbar}\sum_k{
        \langle\vec{P}\cdot\vec{\mu}_{ik}\rangle \left(
            \overbrace{\tilde{E}(t)e^{-i\left(\omega_{ik} - \omega \right) t}e^{-i\vec{k}\cdot\vec{z}}}^{\mathrm{abs.}}	
            + \overbrace{\tilde{E}^*(t)e^{-i\left(\omega_{ik} + \omega \right) t}e^{i\vec{k}\cdot\vec{z}}}^{\mathrm{em.}}
            \right)
        \tilde{\rho}_{kj}}}^{k \leftrightarrow i} \\%{\tilde{\rho}_{ij} \;\mathrm{gain}} \\
        & - \overbrace{\frac{i}{2\hbar}\sum_k{
        \langle\vec{P}\cdot\vec{\mu}_{kj} \rangle
            \left(
            \underbrace{\tilde{E}(t)e^{-i\left(\omega_{kj} + \omega \right)t}e^{i\vec{k}\cdot\vec{z}}}_{\mathrm{em.}}	
            + \underbrace{\tilde{E}^*(t)e^{-i\left(\omega_{kj} - \omega \right)t}e^{-i\vec{k}\cdot\vec{z}}}_{\mathrm{abs.}}
            \right)
        \tilde{\rho}_{ik}}}^{k \leftrightarrow j}%{\tilde{\rho}_{ij} \;\mathrm{loss}}.
\end{aligned}
\end{equation}

We will assume the dipoles are aligned with the electric field here.  The couplings of various coherences and populations with eachother through the electric fields can be visualized by showing the matrix $\mu$.  We now define the matrix and show it's elements (we have arbitrarily assigned $\mu_{21} = 2 \mu_{10}$ for this visualization, so that they are easy to distinguish in the matrix).  

```python
def mu(m10, m21):
    out = np.zeros((9,9))
    # write out only one side of the diagonal
    out[0,1] = m10
    out[0,3] = -m10
    out[1,2] = m21
    out[1,5] = -m10
    out[2,6] = -m10
    out[3,4] = -m21
    out[3,5] = m10
    out[4,7] = m10
    out[5,6] = m21
    out[5,7] = -m21
    out[6,8] = -m21
    out[7,8] = m21
    out = out + out.T
    return out
# illustrate how elements are coupled:
plt.pcolor(mu(1.,2.))
plt.colorbar()
```

We now define the gamma matrix.   On-diagonal elements will indicate the total loss rates.  Off-diagonal elements will only be due to population transfer (if there is no coherence transfer, then we will only include off diagonal elements for relaxation processes, i.e. $11\rightarrow 00$). For now, I will neglect population relaxation (defaults to $\Gamma = 0$ for populations).

```python
# construct the gamma matrix:  give all coherences the same dephasing.  
def G(tau_sqc, tau_dqc, g10=0, g21=0):
    out = np.zeros((9,9))
    gs = 1/float(tau_sqc)
    gd = 1/float(tau_dqc)
    for i in [1,3,6,7]:
        out[i,i] = -gs
    out[2,2] = -gd
    out[4,4] = -gd
    out[5,5] = -g10
    out[0,5] = g10
    out[8,8] = -g21
    out[5,8] = -g21
    return out
print G(50,50)
```

We now define the electric field we will apply on the sample.  Since we are trying to extract a TrEE experiment, We will assume we have three pulses coincident on the sample.  These pulses have a width of 50 time units, and their period is 4 time units.  We will first consider an experiment when w2 and w2' (time=~200) are coincident, and w1 comes in later (time=~500).  If we are thinking ahead to TrEE, this should be pathways 5 and 6.  

When incorporating the electric field into the hamiltonian, we need to keep track of which Euler component is used under the rotating wave approximation.  If we don't do that, we don't specify which interaction uses which euler component upon integration.  Thus, our pulses will be defined as two separate functions:
\begin{equation}
E(t) = \mathcal{E}^+ + \mathcal{E}^- = \tilde{E} e^{i(kz-wt)} + c.c.
\end{equation}

We will also assume our electric fields are resonant--this makes the equations less complicated (the deturning factors in our Liouville equation above will reduce to a time-independent factor of 1).  The caveat here is I cannot use the dipole matrix we defined above and multiply it by the electric fields, because it needs to know which part of the rotating wave we used (this determines which $k$-vector is internalized for each interaction).  For this reason we split the dipole matrix into to parts below; one for the positive $k$ interactions, and one for the negative $k$ interactions.  

We also assume our pulses have simple gaussian envelopes.   

```python
t = np.arange(1000)

def mupm(m10, m21):
    # return two matrices; one for the plus k interactions, and one for the minus k interactions
    out = np.zeros((9,9))
    # write out only one side of the diagonal
    out[0,3] = -m10
    out[1,0] = m10
    out[1,5] = -m10
    out[2,1] = m21
    out[2,6] = -m10
    out[3,4] = -m21
    out[5,3] = m10
    out[5,7] = -m21
    out[6,5] = m21
    out[6,8] = -m21
    out[7,4] = m10
    out[8,7] = m21
    out1 = out
    out2 = out.T
    return out1, out2

def pulse(time, phase):
    # return only the -1j euler component of the phase--the other component can just be evaluated from the first when needed
    # (complex conjugates of eachother)
    #sin = np.sin((t-time)/4. + phase)
    gauss = np.exp(-(t-time)**2 / (2*50.**2)) 
    out = np.exp(-1j*phase) * gauss
    return out

E2  = pulse(200,0)
E2p = pulse(200,0)
E1  = pulse(500,0)

#plt.figure()
#plt.plot(t,E1)
#plt.plot(t,E2)
#plt.plot(t,E2p)
#plt.title('individual pulses')
plt.figure()
plt.plot(t,np.abs(E1 + E2 + E2p))
plt.title('example total efield amplitude')
plt.grid()
```

Of course, the absolute delay (and thus phase) of the pulses varies across the sample.  This spatial variance is responsible for phase-matched output, and will be treated later.  For now, let's construct a hamiltonian so we can simulate the effects of these beams.  

## Part 2:  Integrate the hamiltonian

We will integrate the liouville equation using iterative integration:  $f_{next} = f_{now} + \frac{df_{now}}{dt} * dt$.

```python
def integrate(E, m10, m21, g):
    rho_0 = np.zeros((9))
    rho_0[0] = 1.
    rho = np.zeros((len(t),9), dtype=np.complex)
    #rho[0] = rho_0
    mp, mm = mupm(m10, m21)
    gamma = G(g,g)
    for ti in range(len(t)):
        # need to make sure Ep and Em are used on the right elements:
        Hi = complex(0,1) * (mp * E[ti] + mm * E[ti].conj()) + gamma
        if ti==0: last_rho = rho_0
        else: last_rho = rho[ti-1]
        drho = np.dot(Hi,last_rho)
        rho[ti] = last_rho + drho
    return rho
```

```python
# a simple function used to visualize the different density matrix elements
def plotter(rho):
    labels = ['00','10','20','01','02','11','21','12','22']
    """
    plt.figure()
    for i in range(rho.shape[-1])[1:]:
        plt.plot(np.real(rho[:,i]), label=labels[i])
    plt.grid()
    plt.legend(loc=1, fontsize=10)
    plt.title('real')
    plt.figure()
    for i in range(rho.shape[-1])[1:]:
        plt.plot(np.imag(rho[:,i]), label=labels[i])
    plt.grid()
    plt.legend(loc=1, fontsize=10)
    plt.title('imag')
    """
    plt.figure()
    for i in range(rho.shape[-1]):
        if i in [1,2,5,6,8]:
            plt.plot(np.abs(rho[:,i]), linewidth=5, alpha=0.5, label=labels[i])
    plt.grid()
    plt.legend(loc=1, fontsize=10)
    plt.title('amp')
```

```python
# compare the density matrix propagation with different dipole weights:
test_output1 = integrate((E1+E2+E2p)*1e-3, 1, np.sqrt(2.), 75.)
test_output2 = integrate((E1+E2+E2p)*1e-3, 1, np.sqrt(1.), 75.)
test_output3 = integrate((E1+E2+E2p)*1e-2, 1, np.sqrt(1.), 75.)

plt.figure()
plt.plot(t,np.abs(E1 + E2 + E2p), 'k:', linewidth=5)
plt.grid()
plt.title('total efield amplitude')
plotter(test_output1)
plt.title('harmonic system')
plotter(test_output2)
plt.title('harmonic system with equal overtone/fundamental dipoles')
plotter(test_output3)
plt.title('harmonic system with strong fields')
```

## Part 3:  Extract the polarization

The polarization emitted from the sample is proportional to $\langle \mu \rangle$, which is given by the trace of the matrix $\mu \cdot \rho(t)$.  This expression works when we express our density matrix elements as 3x3 matrix instead of a 9x1 vector.  By computing $Tr\left(\mu \cdot \rho(t)\right)$, we come up with the following weights for our matrix elements (neglecting double quantum coherence, because they are not the relevant frequency that we will isolate):
\begin{equation}
    Tr\left(\mu \cdot \rho(t)\right) = \rho_{10} \mu_{10} - \rho_{01} \mu_{10} + \rho_{21} \mu_{21} - \rho_{12} \mu_{21}
\end{equation}

```python
# ignore the dqc's, since they will be the wrong frequency
def pol(rho, m10, m21):
    dipole = np.zeros((9))
    dipole[1] = float(m10)
    dipole[3] = -float(m10)
    dipole[6] = float(m21)
    dipole[7] = -float(m21)
    p = (rho*complex(0,1)*dipole).sum(axis=-1)
    return p
    
```

```python
# testing a calculation of the polarization:
p1 = pol(test_output1, 1., 1.414)
plt.plot(np.abs(p1), linewidth=2)
plt.grid()
```

## Part 4:  Isolate TrEE polarization

As we mentioned before, the spatial distribution of oscillators and their collective phase relationships will determine the output directions.  Domcke and Fleming realized the phase relationships that produced a polarization in $l\vec{k}_1-m\vec{k}_2+n\vec{k}_3$ direction was the component of the polarization of the sample that had the periodicity of $l\vec{k}_1-m\vec{k}_2+n\vec{k}_3$.  This can be isolated by taking a fourier transform of $P(\phi_1, \phi_2, \phi_3)$.  Fortunately, a very discrete sampling of phases will suffice.  At perturbative fluences and when looking for the phase matching condition $l,m,n = 1,-1,1$ , we need only isolate the DC frequency and the freqency of $\vec{k}$, because higher order frequencies are going to be negligible.  This means we only need 2 samplings per phase variable in our discrete fourier transform (DC and a frequency of $\vec{k}$), for a total of 8 points to sample.  Those 8 points are as follows:

So we will need to calculate $(1) - (2) - (3) - (4) + (5) + (6) + (7) - (8)$.  Due to symmetry of the pulse waveforms in this specific case, we can simplify by observing some equalities ((3) = (4) = -(5) = -(6), (1)=-(8), (2)=-(7)), when we actually do discrete sampling we come up with only three unequal polarization components that need to be calculated.  The appropriate linear combination for these componenets can be expressed as:  $(1)-(2)-2\cdot(3)$.  The different components are visualized below.  You can understand each of these different results as polarizations at different points within the sample.

```python
# 8 possible combinations of constructive and destructive interference, 
# but symmetry allows us to only evaluate half of them explicitly:
x = 1e-3

def make_efield(E1, E2, E2p, phi1, phi2, phi2p):
    e1 = E1 * np.exp(-1j * phi1) 
    e2 = E2 * np.exp(-1j * phi2)
    e2p = E2p * np.exp(-1j * phi2p)
    return e1 + e2 + e2p
plt.figure()
pulses = make_efield(E1, E2, E2p, 0,0,0, )  # (1)
out1 = integrate(pulses*x, 1, 0, 75.) # phi1,phi2,phi2' = 0,0,0 = -(pi,pi,pi)
plt.plot(pulses+pulses.conj(), alpha=0.5, linewidth=5, label=r'$\phi_1, \phi_2, \phi_{2^\prime} = (0,0,0)$')
pulses = make_efield(E1, E2, E2p, np.pi,0,0) # (2)
out2 = integrate(pulses*x, 1, 0, 75.) # pi,0,0 = -(0,pi,pi)
plt.plot(pulses+pulses.conj()+1, alpha=0.5, linewidth=5, label=r'$\phi_1, \phi_2, \phi_{2^\prime} = (\pi,0,0)$')
pulses = make_efield(E1, E2, E2p, 0,np.pi,0) # (3)
out3 = integrate(pulses*x, 1, 0, 75.) # 0,pi,0 = 0,0,pi
plt.plot(pulses+pulses.conj()+2, alpha=0.5, linewidth=5, label=r'$\phi_1, \phi_2, \phi_{2^\prime} = (0,\pi,0)$')
#pulses = make_efield(E1, E2, E2p, np.pi, np.pi, 0)
#out4 = integrate(pulses*x, 1, 0, 75.) # pi,pi,0 = pi,0,pi
#plt.plot(pulses+pulses.conj()+3, alpha=0.5, linewidth=5, label=r'$\phi_1, \phi_2, \phi_{2^\prime} = (\pi,\pi,0)$')
plt.xlim(0, 1200)
plt.legend(loc=0)
plt.title('Pulses')
plt.grid()

#plotter(out2)
p1 = pol(out1, 1., 0)
p2 = pol(out2, 1., 0)
p3 = pol(out3, 1., 0)
#p4 = pol(out4, 1., 0)
plt.figure()
plt.title('total polarization - 2LS')
plt.plot(p1, linewidth=5, alpha=0.5, label=r'$(1)$')
plt.plot(-p2, linewidth=5, alpha=0.5, label=r'$-(2)$')
plt.plot(-2*p3, linewidth=5, alpha=0.5, label=r'$-2\cdot(3)$')
#plt.plot(p4, linewidth=5, alpha=0.5, label=r'$\phi_1, \phi_2, \phi_{2^\prime} = (\pi,\pi,0)$')
plt.xlim(0, 1200)
plt.legend(loc=0)
plt.grid()
```

The parts of the system that did not experience the electric field of E2 and E2' experienced a larger polarization from w1, as can (barely) be seen.  This is due to the fact that areas that interacted with E2 and E2' experienced a degree of population saturation, which inhibited the ability of E1 to interact with the 01 coherence.  The end result of adding these terms together, then will be a net polarization (and it will be the same sign as the dark fringes, since those dominate.  

```python
plt.figure()
plt.title('bright vs. dark fringes - 2LS')
plt.plot(p1-p2, linewidth=5, alpha=0.5, label=r'$(1)-(2)$ (bright)')
plt.plot(-2*p3, linewidth=5, alpha=0.5, label=r'$-2\cdot(3)$ (dark)')
plt.plot(p1-p2 -2*p3, linewidth=5, alpha=0.5, label=r'bright + dark (FWM)')
plt.xlim(0, 1200)
plt.legend(loc=0)
plt.grid()
```

```python
# now let's make a function that does this for us:
def fwmixer(E1, E2, E2p, m10, m21, g, x=1e-3):
    pulses = make_efield(E1, E2, E2p, 0,0,0) # (1)   
    out1 = integrate(pulses*x, m10, m21, g) #
    pulses = make_efield(E1, E2, E2p, np.pi,0,0) # (2)
    out2 = integrate(pulses*x, m10, m21, g) # 
    pulses = make_efield(E1, E2, E2p, 0,np.pi,0) # (3)
    out3 = integrate(pulses*x, m10, m21, g) # 
    p1 = pol(out1, m10, m21)
    p2 = pol(out2, m10, m21)
    p3 = pol(out3, m10, m21)
    fwm = p1 - p2 - 2*p3
    return fwm
```

Now lets consider what output comes out if we change the dipole of the overtone

```python
# we can now compare the effect of having a overtone vs. just a fundamental
fwm1 = fwmixer(E1,E2,E2p, 1, np.sqrt(2.), 75.)
fwm2 = fwmixer(E1,E2,E2p, 1., 0., 75.)
fwm3 = fwmixer(E1,E2,E2p, 1., 1., 75.)
fwm4 = fwmixer(E1,E2,E2p, 1., 0.75, 75.)
plt.plot(fwm1, linewidth=5, alpha=0.5, label=r'$\mu_{21} / \mu_{10} = \sqrt{2}$')
plt.plot(fwm2, linewidth=5, alpha=0.5, label = r'0 (2LS)')
plt.plot(fwm3, linewidth=5, alpha=0.5, label = r'1 (3LS)')
plt.plot(fwm4, linewidth=5, alpha=0.5, label = r'0.75')
plt.grid()
plt.legend(loc=0)
```

I think this is what we expect--harmonic limit gives essentially no signal, and adding the beta pathway with the same dipole as the fundamentals should reduce the signal by half of what it was.   

Well, I guess perturbative expansion of the hamiltonian works out like it should!

One nagging concern about the above result:  why is there some signal in our harmonic case?  Perhaps it's due to ladder climbing by w2 and w2'?  Let's test this by trying lower fluence:  it is ladder climbing if the signal gets ratiometrically smaller:

```python
fwm5 = fwmixer(E1,E2,E2p, 1, np.sqrt(2.), 75., x=1e-4)
fwm6 = fwmixer(E1,E2,E2p, 1., 0., 75., x=1e-4)
plt.plot(fwm5, linewidth=5, alpha=0.5, label=r'$\mu_{21} / \mu_{10} = \sqrt{2}$')
plt.plot(fwm6, linewidth=5, alpha=0.5, label = r'0')
plt.legend(loc=0)
plt.grid()
# indeed it is--so the signal above is an artifact of higher order processes (and also an artifact reflecting we don't
# have the three-quantum level in our basis states)
```

So indeed, we were not at sufficiently low fluence in the above example to call ourselves perturbative; there was some appreciable double quantum excitation by the beams.  Whether that signal is truly what we expect for the TrEE direction was not clear, because we know that since double quantum excitation was non-negligible, our phase cycling routine was not thorough enough, either.  

## Part 5:  Higher order effects and Rabi Cycling

Now let's try to access the higher order effects and use a phase cycling routine that is sensitive to aliasing effects.  The question is, if the higher freqency component of a phase is $n$, how many sampling points to we need so that the frequencies are not aliased?  The answer is $n+1$ points.  We saw that 1e-4 is sufficiently small for 1st order effects to dominate, but that at 1e-3 amplitude, we started deviating.  We probably want to go to ~1e-2 in amplitude to get to observe rabi cycling, so we could presumably need ~100 phase points to sample.  Let's start by defining a new function that can cycle the phases using an arbitrary size Reimann sum to compute the intergral: 

```python
def phase_cycle(E1, E2, E3,
            n1, n2, n3,
            m10, m21, g, x=1e-3):
    # keep E1 perturbative, so no need to extensively phase cycle--in fact, keep it fixed and use relative phase
    out = np.zeros((n1,n2,n3,E1.size), dtype=np.complex)
    for i in range(n1):
        p1 = np.exp(1j*2*np.pi * i / n1)
        for j in range(n2):
            p2 = np.exp(1j*2*np.pi * j / n2)
            for k in range(n3):
                p3 = np.exp(1j*2*np.pi * k / n3)
                outijk = integrate((p1*E1 + p2*E2 + p3*E3)*x,
                                   m10, m21, g)
                out[i,j,k] = pol(outijk, m10, m21)
    out = np.fft.fftn(out, axes=(0,1,2))
    return out
```

We will want to first investigate how many sampling points it takes to separate TrEE from the higher order processes.  We will start test convergence in a couple cases

```python
x = 1e-4
y = 4e2
npts = 20
series1 = []
for i in range(npts)[3:]:
    outi = phase_cycle(E1, E2*y, E2p*y, 
                       3, i, i,
                       1,1, 75, x=x)
    series1.append(np.abs(outi[1,-1,1]).sum())
```

```python
plt.figure()
plt.grid()
x = np.arange(npts)[3:]
plt.plot(x,(series1/x**2 - series1[-1]/(npts-1)**2)/(series1[-1]/(npts-1)**2)*100, linewidth=5)
plt.title('TrEE phase cycling convergence test')
plt.ylabel('% error')
plt.xlabel('number of phase sample points')
```

So it appears even at this rediculously high fluence (20x the fluence that gave us appreciable 2-quantum coherences from a single pulse in the earlier experiment), we need roughly 10-20  phase samples to isolate the signal of interest.  This multiplicative relationship should be expected, given the scalings of susceptabilities.  In fact, higher order diffractions might be suppressed because of our failure to include higher order manifolds than 2 quanta, so the scaling might converge even more quickly.

```python
x = 1e-4
# low fluence but highly sampled in phase
out1 = phase_cycle(E1,E2, E2p, 
                   3, 10, 10,
                   1,1, 75, x=x)
```

```python
# now let's see if we can observe rabi cycling by simply changing input fluence
# this takes a few minutes to run
powers = np.sqrt(np.linspace(0, 1, num=15))*factor
fwms = []

for xi in range(len(powers)):
    # keep E1 perturbative in fluence
    pi = powers[xi]
    sampling = 20
    outi = phase_cycle(E1,E2*pi, E2p*pi, 
                       3, sampling, sampling,
                       1,1, 75, x=x)[1,-1,1]
    fwms.append(np.abs(outi).sum())
```

```python
plt.plot(powers,fwms, linewidth=5, label=r'3LS')
plt.xlabel('E2*E2p (a.u.)')
plt.ylabel('integrated FWM signal')
plt.legend(loc=0)
plt.grid()
```

So it looks like we can model rabi cycling on more complex systems than the two level sysem.  Cool!  I am somewhat surprised, however, that the evolution doesn't look more sinusoidal.  It is clearly lopsided.
